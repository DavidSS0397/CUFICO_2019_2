```
import numpy as np
from random import random 
from scipy import integrate
import matplotlib.pyplot as plt
```

```
N=1000

theta=[]

def funcion(Theta,phi,mu):
    
    pi=np.pi

    constante=1/(phi*(np.sqrt(2*pi)))
    argumento=-(Theta-mu)**2/(2*(phi)**2)

    funcionp=constante*np.exp(argumento)
    return funcionp

phii=0.1
muu=0
```

```
a=-1*np.pi #limite inferior
b=np.pi #limite superior

phi0=(b-a)*random()+a #generador de numeros random 


theta.append(phi0)

for i in range(N):

    phip=(b-a)*random()+a


    deltas=-np.log(funcion(phip,phii,muu)/(funcion(theta[-1],phii,muu)))

    if deltas<0:

        thetai=phip

        theta.append(thetai)

    if deltas>0:

        P=funcion(phip,phii,muu)/(funcion(theta[-1],phii,muu))

        x=random()
        
        if x<P:
            thetai=phip
            theta.append(thetai)

            
print(theta)

plt.hist(theta,bins=100)
plt.show()
```
