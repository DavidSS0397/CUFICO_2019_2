```
import numpy as np
from random import random 
from scipy import integrate
import matplotlib.pyplot as plt
```

```
a=1 #limite inferior
b=10 #limite superior 
c=0
d=2

N=[100,1000,10000]
```

```
def F(x):
    return (x)**-3

def G(x):
    return (x)**-0.5

def H(z,y,x):
    return (x)**-1 +y +z**2

I1=0 #variable para la integral F
I2=0 #variable para la integral G
I3=0

S21=[]
S22=[]
S23=[]
E1=[]
E2=[]
E3=[]
for j in range(len(N)):
    for i in range(N[j]):
   
        x=(b-a)*random() +a 
        y=(d-c)*random() 
        z=(d-c)*random() 
        I1+=F(x)
        I2+=G(x)
        I3+=H(z,y,x)
  

    S21.append(I1**2/N[j]-(I1/N[j])**2)
    S22.append(I2**2/N[j]-(I2/N[j])**2)
    S23.append(I3**2/N[j]-(I3/N[j])**2)
    E1.append(I1/N[j])
    E2.append(I2/N[j])
    E3.append(I3/N[j])

    

    print("con N=", N[j])    
    print("integral F",I1*(b-a)/N[j])
    print("integral G",I2*(b-a)/N[j])
    print("integral H",(I3*(b-a)*(d-c)**2/N[j]))

   
print("integral F",integrate.quad(F,1,10))
print("integral G",integrate.quad(G,1,10))
print("integral H",integrate.tplquad(H,1,10,2,0,2,0,))


fig= plt.figure(figsize=(12,12))

#Grafica 
ax0= fig.add_subplot(131)
ax1= fig.add_subplot(132)
ax2 = fig.add_subplot(133)



ax0.plot(N,E1, "p-", label="NvsE1")
ax0.legend()
ax1.plot(N,E2, "p-", label="NvsE2")
ax1.legend()
ax2.plot(N,E3, "p-", label="NvsE3")
ax2.legend()

fig1= plt.figure(figsize=(12,12))


ax0= fig1.add_subplot(131)
ax1= fig1.add_subplot(132)
ax2 = fig1.add_subplot(133)



ax0.plot(N,S21, "p-", label="NvsS21")
ax0.legend()
ax1.plot(N,S22, "p-", label="NvsS22")
ax1.legend()
ax2.plot(N,S23, "p-", label="NvsS23")
ax2.legend()




plt.show()
```
