### 4. Data

```
BackgroundModel = ROOT.TH1F( 'bkg', 'My background model', 20, 50, 1050 )
```

```
expde = ROOT.TF1("expde","([2]*expo)+[3]",50,1050)
expde.SetParameters(0.0,-0.005,100,0.0);
```

```
BackgroundModel.FillRandom("expde",100000)
```

```
SignalModel = ROOT.TH1F( 'sig', 'My signal model', 20, 50, 1050 )
```

```
signal = ROOT.TF1("signal", "gaus", 50, 1050)
signal.SetParameters(1.0,500.0,50.0)
```

```
SignalModel.FillRandom("signal",1000)
```

```
DataFile = ROOT.TFile("/content/drive/Shared drives/PyROOT_in_Google/Data5.root","read")
```

```
DataHisto=DataFile.Get('data')
```

### 5. Fit

```
from array import array
import numpy as np

n = 15000

# mu and chi^2 arrays
mu = array( 'f' )
chi2 = array( 'f' )

# mu initialization
for i in range(1, n):
  mu.append( 0.0001*i )

for m in mu:
  # Fit calculation
  NHist = BackgroundModel.Clone("nhist")
  NHist.Sumw2()
  NHist.Add(SignalModel, m)

  # chi^2 calculation
  chi = 0
  for i in range(1, NHist.GetNbinsX()):
    exp = NHist.GetBinContent(i)
    obs = DataHisto.GetBinContent(i)
    
    chi += (obs-exp)**2/exp

  chi2.append(chi)

index = chi2.index(min(chi2))

print(f'mu para el que chi^2 es mínimo: mu = {mu[index]}')
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(mu, chi2, 'r')
plt.xlabel(r'$\mu$')
plt.ylabel(r'$\chi^2$')
plt.title(r'$\chi^2 \ vs \ \mu$')
plt.grid()
```
