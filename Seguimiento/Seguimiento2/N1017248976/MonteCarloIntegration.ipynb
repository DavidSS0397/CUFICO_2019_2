```python
import numpy as np
import random as random
from scipy import integrate
import matplotlib.pyplot as plt 
```

```python
N =  [100, 1000, 10000, 100000]
a = 1.
b = 10.
```

```python
def f1(x):
  X=x**(-3)
  return X

def f2(x):
  X=x**(-0.5)
  return X

def f3(x,y,z):
  return x**(-1)+y+z**2
```

```python
I = 0.
O = 0.
F1 = []
A = []
S1 = []
E1 = []
for j in range(len(N)):
  for n in range(N[j]):
    x = (b-a)*np.random.random_sample()+a
    I = I + f1(x)
    F1.append(f1(x))
    A.append(I)
  for  i in range(N[j]):
    E = (A[-1]/N[j])
    O = O + (F1[i]**2-E**2)
  S = (b-a)*O/N[j]
  S1.append(S)
  E = (b-a)*(A[-1]/N[j])
  E1.append(E)
print (E1)
print (S1)
print (integrate.quad(f1,a,b))

plt.plot(N, E1)
plt.show()
plt.plot(N, S1)
```

```python
F2 = []
B = []
S2 = []
E2 = []
for j in range(len(N)):
  for n in range(N[j]):
    x = (b-a)*np.random.random_sample()+a
    I = I + f2(x)
    F2.append(f2(x))
    B.append(I)
  for  i in range(N[j]):
    E = (B[-1]/N[j])
    O = O + (F2[i]**2-E**2)
  S = (b-a)*O/N[j]
  S2.append(S)
  E = (b-a)*(B[-1]/N[j])
  E2.append(E)
print (E2)
print (S2)
print (integrate.quad(f2,a,b))
plt.show()
plt.plot(N, E2)
plt.show()
plt.plot(N, S2)
```

```python
c = 0.
d = 2.
F3 = []
C = []
S3 = []
E3 = []
for j in range(len(N)):
  for n in range(N[j]):
    x = (b-a)*np.random.random_sample()+a
    y = (d-c)*np.random.random_sample()+c
    z = (d-c)*np.random.random_sample()+c
    I = I + f3(x,y,z)
    F3.append(f3(x,y,z))
    C.append(I)
  for  i in range(N[j]):
    E = (C[-1]/N[j])
    O = O + (F3[i]**2-E**2)
  S =(b-a)*(d-c)**2*O/N[j]
  S3.append(S)
  E=(b-a)*(d-c)**2*(C[-1]/N[j])
  E3.append(E)
print (E3)
print (S3)

plt.plot(N, E3)
plt.show()
plt.plot(N, S3)


```
